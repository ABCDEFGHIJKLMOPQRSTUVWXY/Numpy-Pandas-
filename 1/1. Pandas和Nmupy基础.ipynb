## 19. ÊèêÂèñscoreÂàó‰∏≠ÂèØ‰ª•Êï¥Èô§5ÁöÑÊï∞Â≠ó‰ΩçÁΩÆÔºöNumpy.argwhere()

```python
data = {"grammer":['Python', 'C', 'Java', 'GO', 'css', 'SQL', 'PHP', 'Python'],
       "score":[1.0, 2.0, 6.0, 4.0, 5.0, 6.0, 7.0, 10.0]}
df = pd.DataFrame(data)
np.argwhere(np.array(df['score'] % 5 == 0))
```

## 20. Êü•Áúã‰∏§ÂàóÂÄºÁõ∏Á≠âÁöÑË°åÂè∑ÔºöNumpy.where()

- np.where()Áî®Ê≥ïÔºö
 - np.where(condition, x, y)Ôºö # Êª°Ë∂≥Êù°‰ª∂(condition)ÔºåËæìÂá∫xÔºå‰∏çÊª°Ë∂≥ËæìÂá∫y„ÄÇ
 - np.where(condition)ÔºöÂè™ÊúâÊù°‰ª∂ (condition)ÔºåÊ≤°ÊúâxÂíåyÔºåÂàôËæìÂá∫Êª°Ë∂≥Êù°‰ª∂ (Âç≥Èùû0) ÂÖÉÁ¥†ÁöÑÂùêÊ†á (Á≠â‰ª∑‰∫énumpy.nonzero)„ÄÇÂΩìÊù°‰ª∂ÊàêÁ´ãÊó∂ÔºåwhereËøîÂõûÁöÑÊòØÊØè‰∏™Á¨¶ÂêàconditionÊù°‰ª∂ÂÖÉÁ¥†ÁöÑÂùêÊ†á,ËøîÂõûÁöÑÊòØ‰ª•ÂÖÉÁªÑÁöÑÂΩ¢Âºè„ÄÇ

```python
df['cycle'] = [2.0, 3.0, 6.0, 3.0, 5.0, 2.0, 7.0, 5.0]
#ÊñπÊ≥ï‰∏ÄÔºöÂÄºÁõ∏Á≠â
df[df['score'] == df['cycle']].index
 
#ÊñπÊ≥ï‰∫åÔºönp.where()
np.where(df.score == df.cycle)
```

## 21. Êü•ÊâæÁ¨¨‰∏ÄÂàóÁöÑÂ±ÄÈÉ®ÊúÄÂ§ßÂÄº‰ΩçÁΩÆÔºöNumpy.sign()

```python
#ÊÄùË∑ØÔºöÊü•ÊâæÊØîÂÆÉÂâç‰∏Ä‰∏™ÂíåÂêé‰∏Ä‰∏™Êï∞Â≠óÈÉΩÂ§ßÊï∞Â≠ó
tem = np.diff(np.sign(np.diff(df['score'])))
np.where(tem == -2)[0] + 1
```

## 22. ÂØπÁ¨¨‰∫åÂàóËÆ°ÁÆóÁßªÂä®Âπ≥ÂùáÂÄºÔºöNumpy.convolve()

```python
np.convolve(df['cycle'], np.ones(3)/3, mode='valid')
```

## 23. ËÆ°ÁÆó‰∏§ÂàóÊ¨ßÂºèË∑ùÁ¶ªÔºöNumpy.linalg.norm()

```python
#ÊñπÊ≥ï‰∏ÄÔºöÂÖ¨ÂºèÊ≥ï
import math as mt
mt.sqrt(sum((df['score']-df['cycle'])**2))
 
#ÊñπÊ≥ï‰∫åÔºönumpyÂáΩÊï∞
np.linalg.norm(df['score'] - df['cycle'])
```

## 24. Êü•ÊâæÊüêÂàóÊï∞ÊçÆ‰∏≠Á¨¨‰∏âÂ§ßÂÄºÁöÑË°åÂè∑ÔºöNumpy.argsort()

## 25. numpyÂÖ∂‰ªñ‰∏Ä‰∫õÁªüËÆ°Âü∫Á°ÄÂáΩÊï∞
