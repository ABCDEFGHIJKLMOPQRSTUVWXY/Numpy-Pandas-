## 19. æå–scoreåˆ—ä¸­å¯ä»¥æ•´é™¤5çš„æ•°å­—ä½ç½®ï¼šNumpy.argwhere()

```python
data = {"grammer":['Python', 'C', 'Java', 'GO', 'css', 'SQL', 'PHP', 'Python'],
       "score":[1.0, 2.0, 6.0, 4.0, 5.0, 6.0, 7.0, 10.0]}
df = pd.DataFrame(data)
np.argwhere(np.array(df['score'] % 5 == 0))
```

## 20. æŸ¥çœ‹ä¸¤åˆ—å€¼ç›¸ç­‰çš„è¡Œå·ï¼šNumpy.where()

- np.where()ç”¨æ³•ï¼š
 - np.where(condition, x, y)ï¼š # æ»¡è¶³æ¡ä»¶(condition)ï¼Œè¾“å‡ºxï¼Œä¸æ»¡è¶³è¾“å‡ºyã€‚
 - np.where(condition)ï¼šåªæœ‰æ¡ä»¶ (condition)ï¼Œæ²¡æœ‰xå’Œyï¼Œåˆ™è¾“å‡ºæ»¡è¶³æ¡ä»¶ (å³é0) å…ƒç´ çš„åæ ‡ (ç­‰ä»·äºnumpy.nonzero)ã€‚å½“æ¡ä»¶æˆç«‹æ—¶ï¼Œwhereè¿”å›çš„æ˜¯æ¯ä¸ªç¬¦åˆconditionæ¡ä»¶å…ƒç´ çš„åæ ‡,è¿”å›çš„æ˜¯ä»¥å…ƒç»„çš„å½¢å¼ã€‚

```python
df['cycle'] = [2.0, 3.0, 6.0, 3.0, 5.0, 2.0, 7.0, 5.0]
#æ–¹æ³•ä¸€ï¼šå€¼ç›¸ç­‰
df[df['score'] == df['cycle']].index
 
#æ–¹æ³•äºŒï¼šnp.where()
np.where(df.score == df.cycle)
```

## 21. æŸ¥æ‰¾ç¬¬ä¸€åˆ—çš„å±€éƒ¨æœ€å¤§å€¼ä½ç½®ï¼šNumpy.sign()

```python
#æ€è·¯ï¼šæŸ¥æ‰¾æ¯”å®ƒå‰ä¸€ä¸ªå’Œåä¸€ä¸ªæ•°å­—éƒ½å¤§æ•°å­—
tem = np.diff(np.sign(np.diff(df['score'])))
np.where(tem == -2)[0] + 1
```
