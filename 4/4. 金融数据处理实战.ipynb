# 6. Âà†Èô§ÊâÄÊúâÂ≠òÂú®Áº∫Â§±ÂÄºÁöÑË°å

- ËØ≠Ê≥ïÔºö
    - df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
- ÂèÇÊï∞Ôºö
    - axis : {0 or 'index', 1 or 'columns'}, default 0
        - 0, or 'index'ÔºöÂà†Èô§ÂåÖÂê´‰∏¢Â§±ÂÄºÁöÑË°å
        - 1, or 'columns'ÔºöÂà†Èô§ÂåÖÂê´‰∏¢Â§±ÂÄºÁöÑÂàó
        - ÈªòËÆ§‰∏∫0
    - how : {'any', 'all'}, default 'any'
        - 'any': Â¶ÇÊûúÂ≠òÂú®NAÂÄºÔºåÂàôÂà†Èô§ËØ•Ë°åÊàñÂàó
        - 'all': Â¶ÇÊûúÊâÄÊúâÂÄºÈÉΩÊòØNAÔºåÂàôÂà†Èô§ËØ•Ë°åÊàñÂàó
    - thresh: int,‰øùÁïôÂê´Êúâint‰∏™ÈùûÁ©∫ÂÄºÁöÑË°å
    - subset: ÂØπÁâπÂÆöÁöÑÂàóËøõË°åÁº∫Â§±ÂÄºÂà†Èô§Â§ÑÁêÜ

- ‰∏çÂêåÊÉÖÂÜµÂèÇÊï∞ËÆæÁΩÆÔºö
    -  Âà†Èô§ÊâÄÊúâÊúâÁº∫Â§±ÂÄºÁöÑË°å
        - df.dropna()
    - Âà†Èô§ÊâÄÊúâÊúâÁº∫Â§±ÂÄºÁöÑÂàó
        - df.dropna(axis='columns')
        - df.dropna(axis=1)
    - Âà†Èô§ÊâÄÊúâÂÄºÁº∫Â§±ÁöÑË°å
        - df.dropna(how='all')
    - Âà†Èô§Ëá≥Â∞ëÊúâ‰∏§‰∏™ÈùûÁº∫Â§±ÂÄºÁöÑË°å
        - df.dropna(thresh=2)
    - ÊåáÂÆöÂà§Êñ≠Áº∫Â§±ÂÄºÁöÑÂàóËåÉÂõ¥
        - df.dropna(subset=['name', 'born'])
    - ‰ΩøÂà†Èô§ÂíåÁöÑÁªìÊûúÁîüÊïà
        - df.dropna(inplace=True)
    - ÊåáÂÆöÂàóÁöÑÁº∫Â§±ÂÄºÂà†Èô§
        - df.toy.dropna()
